```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
import plotly.express as px
import chart_studio.plotly as py

import ast 
from collections import Counter
from wordcloud import WordCloud
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")

import cufflinks as cf 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.tail(30)
```

```python
credits.head()
```

```python
print("Shape of movies dataset:",movies.shape)
print("Shape of credits dataset:",credits.shape)
```

```python
movies.info()
```

```python
credits.info()
```

```python
# Joining the two datsets
movies = pd.merge(left = movies, right = credits, on='title')
```

```python
movies = movies.drop(columns=['homepage','tagline','id','overview','status','original_title','movie_id'])
```

```python
movies.head(3)
```

```python
movies['genres'][0]
```

```python
# Tidying up genre, production_companies and production_countries column
def func(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies['genres'] = movies['genres'].apply(func)
movies['genres'][0]
```

```python
movies['production_companies'] = movies['production_companies'].apply(func)
movies['production_countries'] = movies['production_countries'].apply(func)
movies.head(3)
```

```python
genres = Counter()
for i in range(movies.shape[0]):
    for j in movies.genres[i]:
        genres[j]+=1
Genres = pd.DataFrame.from_dict(genres, orient='index').reset_index()
Genres = Genres.rename(columns = {'index': 'Genres' ,0: 'Frequency'})
Genres.loc[Genres['Frequency'] < 200, 'Genres'] = 'Others'
```

```python
Genres
```

# Q1: Use bar chart to draw genres of movies.

```python
Genres.plot(kind='bar',x='Genres')
```

# Q2: Use pie chart to draw top 5 languages.

```python
# Top Production Counties
prod_cont = Counter()
for i in range(movies.shape[0]):
    for j in movies.production_countries[i]:
        prod_cont[j]+=1
movie_prod_cont = pd.DataFrame.from_dict(prod_cont, orient='index').reset_index()
movie_prod_cont = movie_prod_cont.rename(columns = {'index': 'Production Country' ,0: 'Frequency'})
movie_prod_cont=movie_prod_cont.sort_values(by = ['Frequency'],ascending=False).reset_index().head(5)
movie_prod_cont
```

```python
movies['spoken_languages'][0]
```

```python
movies['spoken_languages'] = movies['spoken_languages'].apply(func)
movies.head(3)
```

```python
# Top 5 language
lang_cont = Counter()
for i in range(movies.shape[0]):
    for j in movies.spoken_languages[i]:
        lang_cont[j]+=1
movie_lang_cont = pd.DataFrame.from_dict(lang_cont, orient='index').reset_index()
movie_lang_cont = movie_lang_cont.rename(columns = {'index': 'Top 5 languages' ,0: 'Frequency'})
movie_lang_cont=movie_lang_cont.sort_values(by = ['Frequency'],ascending=False).reset_index().head(5)
movie_lang_cont
```

```python


sums = movie_lang_cont.groupby(movie_lang_cont["Top 5 languages"])["Frequency"].sum()
plt.pie(sums, labels=sums.index);
plt.show()
```

# Q3: Use WordCloud to draw genres.

```python

 
comment_words = ''
 
# iterate through the csv file
for val in Genres.Genres:
     
    # typecaste each val to string
    val = str(val)
 
    # split the value
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
 
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()



#or 
#text2 = " ".join(Genres for Genres in Genres.Genres)

# Creating word_cloud with text as argument in .generate() method

#word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text2)

# Display the generated Word Cloud

#plt.imshow(word_cloud, interpolation='bilinear')

#plt.axis("off")

#plt.show()



```
